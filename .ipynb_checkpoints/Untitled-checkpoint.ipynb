# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd #Importing necessary module
```

```python
var = sns.load_dataset("penguins")
var #Load the dataset and see it
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var)
plt.show() #See the simple scatter plot graph
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex")
plt.show() #Differentiate the male aned female category in the graph
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex")
plt.show() #Change the style of the grah plot
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", sizes=(100,40))
plt.show() #Change the plots size
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", palette="gist_ncar")
plt.show() #Change the color of the plots
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", palette="gist_ncar", alpha=0.5)
plt.show() #Decrease the intensity of the plots
```

```python
m = {"Male": "*", "Female": "o"}
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, 
                hue="sex", style="sex", palette="gist_ncar", markers=m)
plt.show() #Change the marker of the plots
```
