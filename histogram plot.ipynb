# Histogram Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd #Importing necessary module
```

```python
var = sns.load_dataset("penguins")
var #Load the dataset and see it
```

```python
sns.displot(var["flipper_length_mm"])
plt.show() #See the simple histogram plot
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230])
plt.show() #Organize the value along the x-axis
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True)
plt.show() #There arise a line within the graph and kde means Krenal Density
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True,rug=True)
plt.show() #there arise some lines thorugh x-axis which indicates kde
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True,rug=True,color="r")
plt.show() #Change the color of the graph
```

```python
sns.displot(var["flipper_length_mm"],kde=True,rug=True,log_scale=True)
plt.show() #It converts the x-axis value into log scale
```
