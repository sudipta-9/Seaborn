```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #Import the necessary libraries
```

```python
var = sns.load_dataset("tips")
var #See the sample dataset of csv file
```

```python
sns.countplot(x="sex",data=var)
plt.show() #See the simple countplot graph
```

```python
sns.countplot(x="sex",data=var,hue="smoker")
plt.show() #Differentiate the male and female in terms of smoker
```

```python
sns.countplot(y="sex",data=var,hue="smoker")
plt.show() #Make the graph horizontally
```

```python
sns.countplot(x="sex", data=var, hue="smoker", palette="bwr")
plt.show() #Add different color palette
```

```python
sns.countplot(x="sex", data=var, hue="smoker", palette="bwr", saturation=0.5)
plt.show() 
```
