```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

y_1 = sns.load_dataset("penguins")

sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",palette="Accent")
plt.show() #Change the color of the graph
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

m = {"Male": "*", "Female": "o"}
y_1 = sns.load_dataset("penguins")

sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",palette="Accent",markers=m)
plt.show() #Set a marker 
```

```python
y_1 = sns.load_dataset("penguins").head(20)
y_1 #Show only top 20 data
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",palette="Accent",markers=["o",">"],dashes=False)
plt.show() #Remove the dash line of line graph
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",palette="Accent",markers=["o",">"],legend=False)
plt.show() #Remove the legend inside the graph
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",palette="Accent",markers=["o",">"],dashes=False)
plt.title("pyton")
plt.grid()
plt.show() #There arise a grid and title in the line graph
```
