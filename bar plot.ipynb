# Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd #Importing necessary module
```

```python
var = sns.load_dataset("penguins")
var #Load the dataset and see it
```

```python
sns.barplot(x=var.island, y=var.bill_length_mm)
plt.show() #See the simplr bar graph according to the datasets
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var)
plt.show() #Another way to see the bar graph
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex")
plt.show() #Differentiate the graph of male and female
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"])
plt.show() #Change the graph order first female and then male
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order1)
plt.show() #Change the order of the label through x-axis
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], errorbar=('ci', 50))
plt.show() #Change the graph size it should be always 0 to 100
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], n_boot=5)
plt.show() #Change the size of the line of the above graph
```

```python
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=var, orient="v")
plt.show() #Change the orientation of the graph vertically
```

```python
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=var, orient="h")
plt.show() #Change the orientation of the graph horizontally
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, color="g")
plt.show() #Change the all graph color green
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], n_boot=5, palette="icefire")
plt.show() #Change the color palette
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], n_boot=5, saturation=0.3)
plt.show() #See the graph color light and the above line color yellow
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], n_boot=5, errcolor="y", errwidth=10)
plt.show() #See the above line color yellow and increase the width size of the line of each graph
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], n_boot=5, errcolor="b", capsize=0.5)
plt.show() #See the samplt cap of the above of the graph
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], n_boot=5, errcolor="b", dodge=False)
plt.show() #Change the graph seperation and seperate the line of the above graph
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], n_boot=5, errcolor="b", alpha=0.5)
plt.show() #Blur the graph
```

```python
sns.set(style="whitegrid")
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"], n_boot=5, errcolor="b")
plt.show() #Set the grid backgroud of the graph
```
